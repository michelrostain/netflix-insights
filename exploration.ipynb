# Netflix DATA analysis    

## AI Watch   

**Intelligence Artificielle :**    
Ensemble des théories et des techniques regroupant des programmes informatiques complexes capables de simuler certains traits de l'intelligence humaine (raisonnement, apprentissage, etc.). Par exemple, un système de recommandation de visionnage pour une plateforme de vidéo en ligne.  

**Machine  Learning :**   
Sous-domaine de l'IA consistant à entraîner des algorithmes à partir de données, afin qu'ils apprennent à effectuer des tâches sans être explicitement programmés pour chacune d'elles.   

**Prétraitement des données :**   
Le prétraitement des données consiste à transformer les données brutes en un format plus adapté et plus pertinent pour l'analyse et l'apprentissage des modèles.   

**Analyse descriptive des données :**   
C'est l'analyse des données passées afin de rendre compte de ce qui s'est produit et cela rends compte de l'état actuel d'un jeu de données. Par exemple, combien de films sont présents dans la base de données, quelle est kla durée moyenne d'un film, y a-t-il des données manquantes, etc.?


### Domaine d'application de l'IA : l'éducation, l'accès à la connaissance et le traitement de celle-ci
**Contexte :**   
L'accès à la connaissance s'applique dans un contexte de rapport profs/élève qui formate la connaissance et l'accès à celle-ci. De ce fait, un sentiment d'exclusion de certains étudiants peut se manifester du fait de ce formatage, et ce même dans une excellence de la transmission de la part du professeur.

**Application :**
L'IA pourrait donc permettre de personnaliser la gestion de la connaissance en fonction de l'élève pour lui permettre de créer un pont entre sa vision (pouvant être interprété par ses lacunes et ses faiblesses dans un système classique) et l'ingestion de la connaissance académique et/ou de terrain. Les questions posées par l'élève, en rapport avec un contenu pédagogique, permet de cerner une manière de réfléchir et ainsi de concevoir un "chemin d'apprentissage" qui dédramatise le rapport à la connaissance pour le rendre plus efficace.
