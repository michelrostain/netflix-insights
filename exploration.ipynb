## AI Watch   

**Intelligence Artificielle :**    
Ensemble des théories et des techniques regroupant des programmes informatiques complexes capables de simuler certains traits de l'intelligence humaine (raisonnement, apprentissage, etc.). Par exemple, un système de recommandation de visionnage pour une plateforme de vidéo en ligne.  

**Machine  Learning :**   
Sous-domaine de l'IA consistant à entraîner des algorithmes à partir de données, afin qu'ils apprennent à effectuer des tâches sans être explicitement programmés pour chacune d'elles.   

**Prétraitement des données :**   
Le prétraitement des données consiste à transformer les données brutes en un format plus adapté et plus pertinent pour l'analyse et l'apprentissage des modèles.   

**Analyse descriptive des données :**   
C'est l'analyse des données passées afin de rendre compte de ce qui s'est produit et cela rends compte de l'état actuel d'un jeu de données. Par exemple, combien de films sont présents dans la base de données, quelle est kla durée moyenne d'un film, y a-t-il des données manquantes, etc.?

### Affichage le nombre de cellules vides par colonnes   
Permet d'afficher le nombre de cellules vides par colonnes. L'argument de sum() est par défaut axis=0 (colonnes). Si on voulait les cellules vides par lignes, il aurait fallu écrire axis=1 en argument de sum() : sum(axis=1)

```python
df.isna().sum()
```

### Afficher le pourcentage de cellules vides par colonne   
Ici affiché avec 4 chiffres après la virgule (arrondi).

```python
print(round(df.isna().sum()*100/len(df), 4))
```
