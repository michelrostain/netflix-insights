## Changer un format de date dans une colonne et l'insrire dans la base de donnée
Le format de la databse est "septmeber 12, 2023". Il faut le changer de "str" à datetime

```python
# Applicable qu'une seule fois, car  la fois d'après str.strip() ne fonctionnera pas sur des dates déjà converties et renverra une erreur
df["date_added"] = pd.to_datetime(df["date_added"].str.strip())
```

## Modifier la variable de durée pour un film   
Il faut afficher "160" au lieu de "160 min".

```python
# On cére une varialble "film" qui comporte les lignes où la colonne "type" contient "Movie"
film = df["type"] == "Movie"
# "Pour un index donné défini par la variable "film", dans la colonne "duration" remplace "min" par une case vide puis supprime ce vide et d'autres vide si il y en a dans cette case."
df.loc[film, "duration"] = df.loc[film, "duration"].str.replace("min", "").str.strip()
```

## Modifier la variable de durée d'une série   
Il faut afficher "2" au lieu de "2 seasons"

```python

serie = df["type"] == "TV Show"
df.loc[serie, "duration"] = df.loc[serie, "duration"].str.replace("Seasons?", "", case=False, regex=True).str.strip()
```

Transformons ces chiffres de chaine de caractère à entier.
'''str.extract(r'(\d+)')''' : 
- **str** : accède à la manipulation de string dans Pands
- **extract** : methode qui extrait une partie d'une chaine selon un motif regex. Elle retourne uniquement ce aqui correspond au motif et ignore le reste 
- **(r'(--)')** : le r devant les parenthèses signifie "raw strings". Cela permet de ne pas  interpréter les caractères spéciaux entre les parenthèses qui suivent. les "'"définissent la chaine de caracère python.
- **(\d+)** : c'est la regex, décomposée  : 
    - \d : matche un chiffre de 0 à 9
    - "+" : un ou plusieurs chiffres consécutifs
    - () : les parenthèses délimitent le groupe de capture et la synthaxe regex

```python
df["duration"] = df["duration"].str.extract(r'(\d+)').astype("Int64")
```

Vérifions que tout est bien convertis. Le code suivant devra renvoyer "int64" si c'est bon, et "object" si la conversion n'a pas été effectuée sur tou le fichier

### Répartition par rating   
Dans la colonne "Rating", des erreurs de saisie provoqué la présence de durée d'oeuvre au lieu de la classification. Afin de ne pas polluer le rendu en graphique, le code extrait de fait les lignes qui ont le mot "min". Il ne reste donc que les lignes correctement remplies : 
- **~df** : le '~' inverse le filtre

```python
# Seaborn
rating = df[~df["rating"].str.contains("min", na=False)]["rating"].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=rating.index, y=rating.values)
plt.title("Nombre d'oeuvre par rating")
plt.xlabel("Rating")
plt.ylabel("Nombre d'oeuvre")
plt.tight_layout()
plt.show()
```
