# Netflix DATA analysis

## AI Watch   

**Intelligence Artificielle :**    
Ensemble des théories et des techniques regroupant des programmes informatiques complexes capables de simuler certains traits de l'intelligence humaine (raisonnement, apprentissage, etc.).   

**Machine  Learning :**   
Utilisation et dévellopement de programme capables d'apprendre et de s'adapter sans suivre des instructions en utilisant des algorithmes et des modèles statistiques pour analyser et tirer des conclusions en fonction des motifs issus d'un ensemble de données.   

**Prétraitement des données :**   
Le prétraitement des données sert consiste à transformer les données brutes en un format plus adapté et plus pertinent pour l'analyse et l'apprentissage des modèles.   

**Analyse descriptive des données :**   
C'est l'analyse des données  passées afin de rendre compte ce qui s'est produit.   

### Domaine d'application de l'IA : l'éducation, l'accès à la connaissance et le traitement de celle-ci
**Contexte :**   
L'accès à la connaissance s'applique dans un contexte de rapport profs/élève qui formate la connaissance et l'accès à celle-ci. De ce fait, un sentiment d'exclusion de certain(e)s étudiant(e)s peut se manifester du fait de ce formatage, et ce même dans une excellence de la transmission de la part du professeur.

**Application :**
L'IA pourrait donc permettre de personnaliser la gestion de la connaissance en fonction de l'élève pour lui permettre de créer un pont entre sa vision (pouvant être interprété par ses lacunes et ses faiblesses dans un système classique) et l'ingestion de la connaissance académique et/ou de terrain. Les questions posées par l'élève, en rapport avec un contenu pédagogique, permet de cerner une manière de réfléchir et ainsi de concevoir un "chemin d'apprentissage" qui dédramatise le rapport à la connaissance pour le rendre plus efficace.
