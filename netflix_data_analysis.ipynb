... Et avec Matplotlib

```python
import matplotlib.pyplot as plt

# Calcule les valeurs manquantes en pourcentage
missing = df.isna().sum() * 100 / len(df)

# Crée le graphique
fig, ax = plt.subplots(figsize=(12, 5))

# Barres des valeurs manquantes
ax.bar(missing.index, 100 - missing, color='steelblue', label='Présentes')

# Barres des valeurs présentes par dessus
ax.bar(missing.index, missing, bottom=100 - missing, color='red', label='Manquantes')

plt.xticks(rotation=45, ha='right')
plt.ylabel('Pourcentage')
plt.title('Répartition des données manquantes par colonne')
plt.legend()
plt.tight_layout()
plt.show()
```

## Changer un format de date dans une colonne et l'insrire dans la base de donnée
Le format de la databse est "spetmeber 12, 2023". Il faut le changer de "str" à datetime

```python
df["date_added"] = pd.to_datetime(df["date_added"].str.strip())
```

## Modifier la variable de durée pour un film   
Il faut affiche "160" au lieu de "160 min"

```python
# EN premier, sélection des films (les séries ne sont pas concernées)
films = df["type"] == "Movie"
# Cible uniquement les lignes où il est écrit "Movie" dans la colonne "type", puis reprend cette synthaxe pour remplacer "min" par rien, puis supprime l'espace après l nombre (.strip()) et enfin transforme ce qu'il reste en entier naturel (.astype(int)
df.loc[film,"duration"] = df.loc[film, "duration"].str.replace("min", "").str.strip().astype(int)
```
